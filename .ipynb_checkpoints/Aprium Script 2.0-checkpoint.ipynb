```python
def school():
    print("How many classes are you taking this upcoming semester?")
    class_num = input("Please enter  your number of courses: ")
    if class_num == "5":
        print( "Overloading?! You must be a robot yourself.")
        print("Is your school portal something you would like to have on your Aprium Dashboard?")
        school_dash = input("Please enter yes or no to the above question: ")
        if school_dash == "yes":
            print("Duley noted!")
            class_info = input("Please enter in your class information or portal links so it can be added to your dashboard: ")
        else:
            print("Good to know!")
        return activities()
    elif class_num == "4":
        print( "Four classes sounds manageable to me.")
        print("Is your school portal something you would like to have on your Aprium Dashboard?")
        school_dash = input("Please enter yes or no to the above question: ")
        if school_dash == "yes":
            print("Duly noted!")
            class_info = input("Please enter in your class information or portal links so it can be added to your dashboard: ")
        else:
            print("Good to know!")
        return activities()
    elif class_num <= "3": 
        print( "You might have too much fun this semester.")
        print("Is your school portal something you would like to have on your Aprium Dashboard?")
        school_dash = input("Please enter yes or no to the above question: ")
        if school_dash == "yes":
            print("Duly noted!")
            class_info = input("Please enter in your class information or portal links so it can be added to your dashboard: ")
        else:
            print("Good to know!")
        return activities()
    else: 
        print("Sorry I did not understand your input, please enter in a number")
        return school()
    
school()
```

```python
def activities():
    print("There is so much to do at school besides the classes!")
    activ = input("Do you partake in any clubs or sports teams (yes or no): ")
    if activ == "yes".upper():
        print ("I wish I could partake in activities outside of the digital world!")
        types = input("Tell me what kinds of activities you are apart of: ")
        print("Those sounds great!")
        print("Do you want to manage these from your Aprium account?")
        activ_dash = input("Please tell me yes or no: ")
        if activ_dash =="yes".upper():
            print("You already seem like someone who knows how to organize well!")
            activ_info = input("Please enter in your extracurriculars information or links: ")
        else: 
            print("Good to know!")
        return platforms()
    elif activ == "no".upper():
        print ("Extracurriculars are a part of college, hopefully you can find one that sparks your interest!")
        return platforms()
    else: 
        print ("Sorry I do not understand your input, please enter in 'yes' or 'no'.")
        return activities()

activities()
```

```python
def closing():
    print("Thank you so much for taking the time to fill out this information for your account! I cannot wait to begin to set it up for you. In the meantime please let me know what you thought of this experience.")
    rating = input("Please give us a rating of 1-5, five being the highest: ")
    if rating == "5":
        print("SO glad to receive that feedback! Thank you again!")
    elif rating == "4":
        print ("Thank you for your rating!")
    elif rating == "3":
        feedback = input("Please tell me where I could have improved the experience for you: ")
    elif rating == "2":
        feedback = input("Please tell me where I could have improved the experience for you: ")
    elif rating == "1":
        feedback = input("So sorry that you did not have a good experience. Please tell me what I could have done differently: ")   
    else: 
        print("Sorry I didn't understand your input, please enter a number 1-5.")
        return closing()
closing()
```
